```python
#Filling 'Unknown' in place of null values
df.Description = df.Description.fillna("unknown")
print(df.isnull().sum())
```

```python
# Aggregate customer data
df_customers = df.groupby('CustomerID').agg({
    'Revenue': 'sum',  # Total spending
    'InvoiceNo': 'count'  # Number of purchases # Changed 'TransactionID' to 'InvoiceNo'
}).reset_index()

# Rename columns for clarity
df_customers.rename(columns={'Revenue': 'Total_Spend', 'InvoiceNo': 'Purchase_Frequency'}, inplace=True)
# Changed 'TransactionID' to 'InvoiceNo'
```

```python
# Merge the aggregated data (df_customers) with the original DataFrame (df)
# using 'CustomerID' as the key and performing a left merge.
# This adds the aggregated columns (Total_Spend, Purchase_Frequency)
# to the original DataFrame.
df = pd.merge(df, df_customers, on='CustomerID', how='left')

# Now, scale the features 'Total_Spend' and 'Purchase_Frequency'
# using StandardScaler. This is often done in preprocessing for
# machine learning algorithms like K-Means to ensure that features
# with different scales contribute equally to the distance calculations.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['Total_Spend', 'Purchase_Frequency']])
```

**K-Means Clustering**

K-Means is a popular choice for customer segmentation. First, determine the optimal number of clusters using the Elbow Method.

```python
# Importing necessary libraries
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans  # Import KMeans for clustering

# Initialize an empty list to store WCSS values
wcss = []

# Iterate through different cluster numbers (1 to 10)
for i in range(1, 11):
    # Create a KMeans model with 'i' clusters, random state for reproducibility, and 10 initializations
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    # Fit the KMeans model to the scaled features
    kmeans.fit(scaled_features)
    # Append the within-cluster sum of squares (inertia) to the list
    wcss.append(kmeans.inertia_)

# Create a plot to visualize the Elbow method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')  # Plot WCSS values against cluster numbers
plt.xlabel('Number of Clusters')  # Label the x-axis
plt.ylabel('WCSS')  # Label the y-axis
plt.title('Elbow Method for Optimal K')  # Set the title of the plot
plt.show()  # Display the plot
```

**Applying K-Means Clustering**

```python
# Apply K-Means with optimal clusters (assume K=3)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Analyze segment statistics
cluster_summary = df.groupby('Cluster').agg({'Total_Spend': 'mean', 'Purchase_Frequency': 'mean', 'CustomerID': 'count'}).rename(columns={'CustomerID': 'Customer_Count'})
print(cluster_summary)
```

**Estimating Customer Lifetime Value (CLV)**:

CLV helps predict the total revenue a business can expect from a customer over their entire relationship. Below are two common approaches

```python
import datetime as dt

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
latest_date = df['InvoiceDate'].max()  # Most recent transaction date

# Calculate RFM metrics
# Changed 'TransactionID' to 'InvoiceNo' as it's the actual column name in your DataFrame
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency
    'InvoiceNo': 'count',  # Frequency
    'Revenue': 'sum'  # Monetary
}).reset_index()

# Rename columns
rfm.rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'Revenue': 'Monetary'}, inplace=True) # Changed 'TransactionID' to 'InvoiceNo'

# Estimate CLV (Simple method: Avg Monetary * Frequency)
rfm['CLV'] = rfm['Monetary'] * rfm['Frequency']

print(rfm.head())  # Check RFM and CLV values
```
